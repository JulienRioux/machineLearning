```python
%matplotlib inline 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# Load the training set and the test one
from sklearn.datasets import load_boston
boston = load_boston()

# Create the pandas df
boston_df = pd.DataFrame(boston.data)

for feature in range(len(boston.feature_names)):
    boston.feature_names[feature] = boston.feature_names[feature].lower()

# Set the right names to the columns
boston_df.columns = boston.feature_names

# set the medv price column name
boston_df['medv'] = boston.target

boston_df.head()
```

```python
boston_df.describe()
```

```python
# Set the X and y values
X = boston_df
y = boston_df.pop("medv")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 5)
```

```python
# Import the class
from sklearn.linear_model import LinearRegression

# Instantiate the model (default parameters)
linearReg = LinearRegression()

# fit the model with data
linearReg.fit(X_train, y_train)

# predict the response for new observations
y_pred = linearReg.predict(X_test)

```

```python
y_pred
```

```python
mse = sklearn.metrics.mean_squared_error(y_test, y_pred)
print("The Mean Squared Error is:", mse)
```

```python
# Calculate the prediction error for every prediction
errorPred = abs(y_test-y_pred)
```

```python
# Plot the boston house price prediction
fig = plt.figure(dpi=200)
with sns.axes_style("darkgrid", {"axes.facecolor": ".2", 'grid.color': '.18' }):
    cmap = sns.cubehelix_palette(start=1.8, light=0.71, dark=0.26, rot=0.61, hue=-10, as_cmap=True)
    ax = plt.scatter(y_test, y_pred, s=errorPred, cmap=cmap, c=errorPred, alpha=0.9)
    plt.xlabel("Prices")
    plt.ylabel("Predicted Prices")
    plt.title("Boston House Price Prediction", fontsize=18)

    
fig.colorbar(ax)

plt.show()
```
