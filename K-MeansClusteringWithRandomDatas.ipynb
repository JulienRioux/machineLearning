# K Means Clustering with Randoms Datas

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style("darkgrid")
```

```python
# Generate isotropic Gaussian blobs for clustering.
from sklearn.datasets import make_blobs
```

```python
# Create Data
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)
```

## Visualize the data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='Dark2')
```

## KMeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
# KMeans is unsupervised so we put all the datas in the fit method
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='Dark2')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='Dark2')
```
