```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Load the data
from sklearn import datasets
digits = datasets.load_digits()
```

```python
digits
```

```python
digits.target
```

```python
# Check some of the the images and the labels
plt.figure(figsize=(40,8))
sns.set_style("white")

row = 0
col = 0
for index, (image, label) in enumerate(zip(digits.data[0:10], digits.target[0:10])):
    plt.subplot(1, 10, index + 1)
    
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize = 20)

plt.show()
```

```python
# Set the training and test datas
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3, random_state=20)
```

```python
# Choosing the logisticr
from sklearn.linear_model import LogisticRegression

logReg = LogisticRegression()

# Train the model 
logReg.fit(X_train, y_train)

# Test the model
y_pred = logReg.predict(X_test)

score = logReg.score(X_test, y_test)

print("Accuracy:", score)
```

```python
# Calculate the prediction error for every prediction
errorPred = 1/abs(y_test-y_pred+1)
```

```python
# Plot the boston house price prediction
fig = plt.figure(dpi=200)

sns.kdeplot(y_test, y_pred, shade=True)

plt.scatter(y_test, y_pred, s=errorPred*50, color="#003d00", alpha=0.2)

plt.axis("square")
plt.axis("off")


plt.ylabel("Predictions")


plt.show()
```

```python
from sklearn import metrics
```

```python
countMatrix = metrics.confusion_matrix(y_pred, y_test)

countMatrix
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(countMatrix, annot=True, fmt=".3f", linewidth=0.5, square = True, cmap = "Greens_r")

plt.xlabel("Test Digits")
plt.ylabel("Predicted Digits")
all_sample_title = "Accuracy Score: {}".format(score)
plt.title(all_sample_title, size=15)

plt.show()
```
