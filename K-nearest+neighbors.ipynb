```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
sns.set_style("darkgrid")
```

```python
df = pd.read_csv("Classified Data", index_col=0)
```

```python
df.head()
```

# Scale our datas

```python
# Scale our datas
from sklearn.preprocessing import StandardScaler
```

```python
# Instantiate the model
scaler = StandardScaler()
```

```python
# Fit only to the datas not to the target
scaler.fit(df.drop("TARGET CLASS", axis=1))
```

```python
# Perform standardization by centering and scaling
scaled_features = scaler.transform(df.drop("TARGET CLASS", axis=1))
```

```python
scaled_features
```

```python
df.columns
```

```python
# Recreate a scaled version of our features in a dataframe
df_feat = pd.DataFrame(data=scaled_features, columns=df.columns[:-1], )
```

```python
df_feat.head()
```

# Train and predict

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_feat
y = df["TARGET CLASS"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Start with k=1

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Instantiate
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
# Find the best k for the KNeighbor
error_rate = []

for i in range(1, 100):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
# plot the error rate for every k
plt.figure(dpi=200)
plt.plot(range(1,100), error_rate, color="b", ls="--", marker="o", markerfacecolor="r", markeredgecolor="r", markersize=4, alpha=0.7)
plt.title("Error Rate vs K value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```

Plot for 20 to 40

```python
# Find the best k for the KNeighbor
error_rate = []


for i in range(20, 60):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
# plot the error rate for every k
plt.figure(dpi=200)
plt.plot(range(20,60), error_rate, color="b", ls="--", marker="o", markerfacecolor="r", markeredgecolor="r", markersize=7, alpha=0.7)
plt.title("Error Rate vs K value")
plt.xlabel("K")
plt.ylabel("Error Rate")
```
